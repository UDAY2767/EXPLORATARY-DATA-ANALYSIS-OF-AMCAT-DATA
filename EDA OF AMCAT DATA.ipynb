### ** Step -1 Introduction**

The dataset under analysis was released by **Aspiring Minds** as part of the **Aspiring Mind Employment Outcome 2015 (AMEO)** study, which primarily focuses on engineering graduates. The study aims to understand the employment outcomes of students in engineering disciplines and assess the influence of various factors on their job prospects and salary outcomes. This dataset is crucial for evaluating factors such as cognitive abilities, technical skills, and personality traits on the employment results of graduates.

The dataset contains approximately 4,000 records with 40 variables, including demographic features, academic performance, cognitive skill assessments, and job-related information. The target variable for this analysis is **Salary**, which represents the annual cost to company (CTC) offered to the candidates in INR.

The objective of this exploratory data analysis (EDA) is to:
1. Investigate the distribution of various features and understand their relationships with the target variable (Salary).
2. Identify patterns between categorical and continuous variables.
3. Explore the relationships between skills, academic background, and employment outcomes.
4. Answer specific research questions, such as the influence of certain job roles on salary and whether gender affects the choice of specialization.

#### **Dataset Overview**

The dataset consists of the following major features:

| **Variable**           | **Type**              | **Description**                                                  |
|------------------------|-----------------------|------------------------------------------------------------------|
| **ID**                 | UID                   | A unique identifier for each candidate.                          |
| **Salary**             | Continuous            | The annual CTC offered to the candidate in INR.                  |
| **DOJ**                | Date                  | Date of joining the company.                                     |
| **DOL**                | Date                  | Date of leaving the company.                                     |
| **Designation**        | Categorical           | The job title or designation offered in the job.                 |
| **JobCity**            | Categorical           | Location of the job (city).                                      |
| **Gender**             | Categorical           | Gender of the candidate.                                         |
| **DOB**                | Date                  | Date of birth of the candidate.                                  |
| **10percentage**       | Continuous            | Overall percentage obtained in grade 10.                         |
| **12percentage**       | Continuous            | Overall percentage obtained in grade 12.                         |
| **CollegeGPA**         | Continuous            | Aggregate GPA obtained during graduation.                        |
| **Specialization**     | Categorical           | The specialization pursued during the candidate's degree.        |
| **English, Logical, Quant** | Continuous      | Standardized scores in AMCAT sections for cognitive skills.      |
| **Personality Traits**  | Continuous/Standardized | Scores in AMCAT personality assessments (e.g., conscientiousness, extraversion, etc.). |

The dataset offers a comprehensive view of the various factors contributing to the employment outcomes of engineering graduates, with a particular emphasis on **Salary** as the dependent variable.

```python
# Univariate Analysis for Numerical Columns
numerical_columns = ['Salary', '10percentage', '12percentage', 'collegeGPA', 'English', 'Logical', 'Quant', 'Domain']

for column in numerical_columns:
    plt.figure(figsize=(12, 6))
    
    # Histogram with KDE (Probability Density Function)
    plt.subplot(1, 2, 1)
    sns.histplot(data[column], kde=True, bins=20, color='skyblue')
    plt.title(f"Histogram and PDF of {column}", fontsize=16)
    plt.xlabel(column, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    
    # Boxplot for Outlier Detection
    plt.subplot(1, 2, 2)
    sns.boxplot(x=data[column], color='lightgreen')
    plt.title(f"Boxplot of {column}", fontsize=16)
    plt.xlabel(column, fontsize=14)
    
    # Adjust layout for clarity
    plt.tight_layout()
    plt.show()

    # Observations for Numerical Columns
    print(f"Observations for {column}:")
    print(f"Mean: {data[column].mean():.2f}, Median: {data[column].median():.2f}, Std Dev: {data[column].std():.2f}")
    print(f"Minimum: {data[column].min()}, Maximum: {data[column].max()}")
    print(f"Number of missing values: {data[column].isnull().sum()}")
    print("\n")

# Univariate Analysis for Categorical Columns
categorical_columns = ['Designation', 'JobCity', 'Gender', 'Specialization', 'CollegeTier']

for column in categorical_columns:
    plt.figure(figsize=(10, 5))
    
    # Countplot for Categorical Variables
    sns.countplot(y=column, data=data, palette='Set2', order=data[column].value_counts().index)
    plt.title(f"Countplot of {column}", fontsize=16)
    plt.ylabel(column, fontsize=14)
    plt.xlabel('Count', fontsize=14)
    
    # Show plot
    plt.tight_layout()
    plt.show()

    # Observations for Categorical Columns
    print(f"Observations for {column}:")
    print(f"Unique Values: {data[column].nunique()}")
    print(f"Most common value: {data[column].mode()[0]}")
    print(f"Number of missing values: {data[column].isnull().sum()}")
    print("\n")
```

```python
# Function to create a clear countplot with annotations (top 10)
def clear_countplot(column, data):
    plt.figure(figsize=(12, 6))  # Adjust figure size for clarity
    top_10 = data[column].value_counts().nlargest(10).index  # Get top 10 values
    ax = sns.countplot(y=column, data=data[data[column].isin(top_10)], 
                       palette='Set2', order=top_10)
    plt.title(f"Top 10 Countplot of {column}", fontsize=16)
    plt.ylabel(column, fontsize=14)
    plt.xlabel('Count', fontsize=14)

    # Adding count annotations on the bars
    for p in ax.patches:
        ax.annotate(f'{p.get_width()}', 
                    (p.get_width(), p.get_y() + p.get_height() / 2), 
                    ha='left', va='center', fontsize=12)

    plt.tight_layout()  # Adjust layout for better display
    plt.show()

# Categorical Columns
categorical_columns = ['Specialization']

for column in categorical_columns:
    clear_countplot(column, data)

# Observations for the Categorical Column (Specialization)
print(f"Observations for {categorical_columns[0]}:")
print(f"Unique Values: {data[categorical_columns[0]].nunique()}")
print(f"Most common value: {data[categorical_columns[0]].mode()[0]}")
print(f"Number of missing values: {data[categorical_columns[0]].isnull().sum()}")

# Print top 10 values
print("\nTop 10 most common values for Specialization:")
print(data[categorical_columns[0]].value_counts().nlargest(10))
```

# Step 5 - Research Questions
In this step, we will address two key research questions based on the dataset.

Research Question 1: Testing Salary Claims for Computer Science Engineering Graduates
To test the claim from the Times of India article dated January 18, 2019, stating that fresh graduates in roles such as Programming Analyst, Software Engineer, Hardware Engineer, and Associate Engineer can earn between 2.5 to 3 lakhs, we will analyze the salary data for these roles.

A. Filtering the Data for Relevant Roles
We'll first filter the dataset to include only those job titles mentioned in the claim and then check the average salaries for these roles.

```python
# Display unique values in the Designation column
print(data['Designation'].unique())
```

```python
# Define a function to check for substring matches
def check_roles(role):
    return any(sub in role.lower() for sub in ['programming analyst', 'software engineer', 'hardware engineer', 'associate engineer'])

# Filter the dataset for roles that contain relevant substrings
filtered_data = data[data['Designation'].apply(check_roles)]

# Remove entries with missing salary
filtered_data = filtered_data[filtered_data['Salary'].notnull()]

# Calculate average salary for these roles
if not filtered_data.empty:
    average_salary = filtered_data['Salary'].mean()
else:
    average_salary = None

salary_range = (2.5 * 100000, 3 * 100000)  # Convert lakhs to actual salary values

if average_salary is not None:
    print(f"Average Salary for selected roles: {average_salary:.2f} INR")
    print(f"Claimed Salary Range: {salary_range[0]} - {salary_range[1]} INR")
    
    # Observations
    if salary_range[0] <= average_salary <= salary_range[1]:
        print("The claim is supported by the data.")
    else:
        print("The claim is not supported by the data.")
else:
    print("No valid salary data available for the specified roles.")
```

# Research Question 2: 
    - Is there a relationship between gender and specialization? (i.e. Does the preference of Specialisation depend on the Gender?)
A. Creating a Stacked Bar Plot for Gender and Specialization

```python
# Group data by Gender and Specialization
specialization_gender = data.groupby(['Gender', 'Specialization']).size().unstack()

# Plotting the stacked bar chart
specialization_gender.plot(kind='bar', stacked=True, figsize=(12, 6), color=['#66c2a5','#fc8d62','#8da0cb'])
plt.title('Stacked Bar Plot of Gender by Specialization')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Specialization')
plt.grid()
plt.tight_layout()
plt.show()

# Observations
print("Observations on Gender and Specialization:")
for gender in specialization_gender.index:
    print(f"{gender}:")
    for specialization in specialization_gender.columns:
        count = specialization_gender.loc[gender, specialization] if specialization in specialization_gender.columns else 0
        print(f"  {specialization}: {count} candidates")
```

# 2. Impact of College Tier on Salaries

```python
# Check the columns in the dataset
print(data.columns)
```

```python
# Boxplot to compare salaries between Tier 1 and Tier 2 colleges
plt.figure(figsize=(8, 6))
sns.boxplot(x='CollegeTier', y='Salary', data=data)
plt.title('Impact of College Tier on Salaries')
plt.xticks(ticks=[0, 1], labels=['Tier 1', 'Tier 2'])
plt.xlabel('College Tier')
plt.ylabel('Salary (INR)')
plt.show()

# Perform statistical significance test (e.g., t-test)
from scipy import stats

tier1_salaries = data[data['CollegeTier'] == 1]['Salary']
tier2_salaries = data[data['CollegeTier'] == 2]['Salary']

t_stat, p_value = stats.ttest_ind(tier1_salaries, tier2_salaries, nan_policy='omit')

print(f'T-test statistics: {t_stat}, P-value: {p_value}')
```
